```python
from IPython.display import Audio, display
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
from scipy.signal import get_window
```

### A1. Use numpy.fft.fft() to transform the speech signal to its spectral domain. Please plot the amplitude part
### of the spectral components and observe it. Use numpy.fft.ifft() to inverse transform the frequency spectrum
### to time domain signal.  

```python
# Load the speech signal
y, sr = librosa.load("C:\\Users\\Tanya Simhadri\\Downloads\\AI_in_Speech_Processing_recording.mpeg")

# Perform FFT on the speech signal
y_fft = np.fft.fft(y)

# Calculate the amplitude
amplitude = np.abs(y_fft)

# Plot the original signal
plt.figure(figsize=(10, 4))
plt.plot(y)
plt.title('Original Signal')
plt.show()

# Plot the amplitude part
plt.figure(figsize=(10, 4))
plt.plot(amplitude)
plt.title('Amplitude Spectrum of Speech Signal')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.show()

# Inverse transform the frequency spectrum
y_ifft = np.fft.ifft(y_fft)

# Plot the signal after inverse transformation
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y_ifft.real, sr=sr)
plt.title('Time Domain Signal after Inverse Transform')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

### A2. Use a rectangular window to select the low frequency components from your spectrum. Inverse 
### transform the filtered spectrum and listen to this sound. Repeat the same for band pass and high pass
### frequencies of spectrum. 

```python
# Rectangular or Boxcar window

# Define the frequency components
frequency_bins = np.fft.fftfreq(len(y), 1/sr)

# LOW PASS FILTERING
# Apply the Boxcar window
boxcar_window = get_window(('boxcar',), len(y))
filtered_spectrum_low_boxcar = y_fft * boxcar_window
filtered_signal_low_boxcar = np.fft.ifft(filtered_spectrum_low_boxcar).real

# Plot the signal after low-pass filtering
plt.figure(figsize=(10, 4))
librosa.display.waveshow(filtered_signal_low_boxcar, sr=sr)
plt.title('Low-pass Filtering (Boxcar window)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

# Play the filtered low-pass signal
audio_low_pass_boxcar = Audio(filtered_signal_low_boxcar, rate=sr)
display(audio_low_pass_boxcar)

# BANDPASS FILTERING
# Apply the Boxcar window
boxcar_window_bandpass = get_window(('boxcar',), len(y))
bandpass_mask_boxcar = (frequency_bins >= 500) & (frequency_bins <= 1500)
filtered_spectrum_bandpass_boxcar = y_fft * bandpass_mask_boxcar * boxcar_window_bandpass
filtered_signal_bandpass_boxcar = np.fft.ifft(filtered_spectrum_bandpass_boxcar).real

# Plot the signal after bandpass filtering
plt.figure(figsize=(10, 4))
librosa.display.waveshow(filtered_signal_bandpass_boxcar, sr=sr)
plt.title('Bandpass Filtering (Boxcar window)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

# Play the filtered bandpass signal
audio_bandpass_boxcar = Audio(filtered_signal_bandpass_boxcar, rate=sr)
display(audio_bandpass_boxcar)

# HIGH-PASS FILTERING
# Apply the Boxcar window
boxcar_window_highpass = get_window(('boxcar',), len(y))
highpass_mask_boxcar = (frequency_bins > 1500)
filtered_spectrum_highpass_boxcar = y_fft * highpass_mask_boxcar * boxcar_window_highpass
filtered_signal_highpass_boxcar = np.fft.ifft(filtered_spectrum_highpass_boxcar).real

# Plot the signal after high-pass filtering
plt.figure(figsize=(10, 4))
librosa.display.waveshow(filtered_signal_highpass_boxcar, sr=sr)
plt.title('High-pass Filtering (Boxcar window)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

# Play the filtered high-pass signal
audio_high_pass_boxcar = Audio(filtered_signal_highpass_boxcar, rate=sr)
display(audio_high_pass_boxcar)
```
