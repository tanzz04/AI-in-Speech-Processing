```python
import librosa
from IPython.display import Audio, display
filename = "C:\\Users\\Tanya Simhadri\\Downloads\\AI_in_Speech_Processing_recording.mpeg"
y, sr = librosa.load(filename)
```

```python
import numpy as np

first_derivative = np.array([y[1] - 2*y[0]])
for i in range(1,len(y)-1):
    first_derivative = np.append(first_derivative,y[i-1] + y[i+1] - 2*y[i])
    
first_derivative
```

```python
# Original signal
librosa.display.waveshow(y)
```

```python
# First derivative using finite difference method
librosa.display.waveshow(first_derivative)
```

# 2

```python
import librosa
import matplotlib.pyplot as plt

# Find zero-crossing points
zero_crossings = []
for i in range(1, len(first_derivative)):
    if (first_derivative[i-1] >= 0 and first_derivative[i] < 0) or (first_derivative[i-1] < 0 and first_derivative[i] >= 0):
        zero_crossing = i - first_derivative[i-1] / (first_derivative[i] - first_derivative[i-1])
        zero_crossings.append(zero_crossing)

# Plot the zero-crossing points
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.scatter(zero_crossings, [y[int(zc)] for zc in zero_crossings], label='Zero Crossing')
plt.title('Original Signal with Zero Crossings')
plt.xlabel('Sample Index')
plt.show()
```
